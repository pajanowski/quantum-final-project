```python
def get_pi_perm(n):
    pi = {}
    set = []
    for i in range(n):
        set.append(i)

    set_len = int(len(set) / 2)
    for i in range(set_len):
        key = random.choice(set)
        set.remove(key)
        value = random.choice(set)
        set.remove(value)
        pi[key] = value
        pi[value] = key
    return pi

# N integer
def get_L(N):
    ret = []
    pi_perm = get_pi_perm(N)
    for i in range(len(pi_perm.keys())):
        ret.append(pi_perm.get(i))
    return ret

```

```python
def reflection(qc):
    qc.h(0)
    qc.h(1)
    qc.h(2)
    qc.h(3)
    qc.x(0)
    qc.x(1)
    qc.x(2)
    qc.x(3)

    qc.cu1(pi/4, 0, 3)
    qc.cx(0, 1)
    qc.cu1(-pi/4, 1, 3)
    qc.cx(0, 1)
    qc.cu1(pi/4, 1, 3)
    qc.cx(1, 2)
    qc.cu1(-pi/4, 2, 3)
    qc.cx(0, 2)
    qc.cu1(pi/4, 2, 3)
    qc.cx(1, 2)

    qc.cu1(-pi/4, 2, 3)
    qc.cx(0, 2)
    qc.cu1(pi/4, 2, 3)

    qc.x(0)
    qc.x(1)
    qc.x(2)
    qc.x(3)
    qc.h(0)
    qc.h(1)
    qc.h(2)
    qc.h(3)
```
